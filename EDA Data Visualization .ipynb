```python
#Relationship between emotions and overall 
import numpy as np
from bokeh.plotting import figure
from bokeh.io import show

#the data
x=BlendDF['AvgJoy'].sample(n= 50000)
y=BlendDF['overall'].sample(n=50000)

# determine best fit line
par = np.polyfit(x, y, 1, full=True)
slope=par[0][0]
intercept=par[0][1]
y_predicted = [slope*i + intercept  for i in x]

#the data 
x1 = BlendDF['AvgSadness'].sample(n= 50000)
y1=BlendDF['overall'].sample(n=50000)
par1 = np.polyfit(x1, y, 1, full=True)
slope1=par1[0][0]
intercept1=par1[0][1]
y_predicted1 = [slope*i + intercept  for i in x1]

#the data
x2 = BlendDF['AvgDisgust'].sample(n= 50000)
y2=BlendDF['overall'].sample(n=50000)
par2 = np.polyfit(x2, y, 1, full=True)
slope2=par2[0][0]
intercept2=par2[0][1]
y_predicted2 = [slope*i + intercept  for i in x2]

# plot it
fig=figure()
fig.line(x,y_predicted,color='red',legend='y='+str(round(slope,2))+'x+'+str(round(intercept,2)))
fig1=figure()
fig1.line(x1,y_predicted1,color='blue',legend='y='+str(round(slope1,2))+'x+'+str(round(intercept1,2)))
fig2=figure()
fig2.line(x2,y_predicted2,color='yellow',legend='y='+str(round(slope2,2))+'x+'+str(round(intercept2,2)))
show(fig)
show(fig1)
show(fig2)
```

```python
import numpy as np
from bokeh.plotting import figure
from bokeh.io import show

#the data
x=BlendDF['AvgJoy'].sample(n=50000)
y=BlendDF['overall'].sample(n=50000)

# determine best fit line
par = np.polyfit(x, y, 1, full=True)
slope=par[0][0]
intercept=par[0][1]
y_predicted = [slope*i + intercept  for i in x]

# plot it
fig=figure()
fig.line(x,y_predicted,color='red',legend='y='+str(round(slope,2))+'x+'+str(round(intercept,2)))
show(fig)
```
